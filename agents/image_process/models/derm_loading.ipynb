```python
import os
from dotenv import load_dotenv
from pathlib import Path
from huggingface_hub import snapshot_download
```

```python
load_dotenv()
hf_token = os.getenv("HF_TOKEN")
```

```python
PROJECT_ROOT = Path.cwd().parents[2]
DATA_DIR = PROJECT_ROOT / "agents" / "image_process" / "data" / "outputs"
MODEL_DIR = DATA_DIR / "models"
MODEL_DIR.mkdir(parents=True, exist_ok=True) 
```

```python
os.environ["HF_HUB_DISABLE_XET"] = "1"
```

```python
downloaded_path = snapshot_download("google/derm-foundation", token=hf_token, cache_dir=str(MODEL_DIR),local_dir_use_symlinks=False)
print("Downloaded to:", downloaded_path)
```
