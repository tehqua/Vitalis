```python
from transformers import pipeline
import torch
from IPython.display import display, Markdown
```

```python
# Model path
downloaded_path = "./google_medgemma_4b_it"

# Initialize pipeline
pipe = pipeline(
    "text-generation",
    model=downloaded_path,
    device="cuda",
)

# Prompt
prompt = "How do you differentiate bacterial from viral pneumonia?"
role_instruction = "You are a helpful medical assistant."
system_instruction = role_instruction

max_new_tokens = 300

# Message structure
messages = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": system_instruction
            }
        ],
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": prompt
            }
        ],
    },
]

# Generate output
output = pipe(
    messages,
    max_new_tokens=max_new_tokens,
    do_sample=False
)

response = output[0]["generated_text"][-1]["content"]
```

```python
display(Markdown(f"---\n\n**[ User ]**\n\n{prompt}\n\n---"))
display(Markdown(f"**[ MedGemma ]**\n\n{response}\n\n---"))
```
